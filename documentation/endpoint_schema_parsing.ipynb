```python
import json

def print_json_attributes(json_string, prefix=""):
  data = json.loads(json_string)
  for key, value in data.items():
    if isinstance(value, dict):
      print_json_attributes(json.dumps(value), prefix + key + ".")
    elif isinstance(value, list):
      for i, item in enumerate(value):
        if isinstance(item, dict):
          print_json_attributes(json.dumps(item), prefix + key + "[ ].")
        else:
          print(prefix + key + "[ ]")
    else:
      print(prefix + key)

# Example usage
#filename = 'KodeforType_schema.json'
#filename = 'KodeforGrunntype_schema.json'
#filename = 'KodeforKartleggingsenhet_schema.json'
#filename = 'variabler_allekoder_schema.json'
#filename = 'kodeforVariabel_schema.json'
#filename = 'kodeforVariabelnavn_schema.json'
#filename = 'maaleskala_schema.json'
filename = 'type_allekoder_schema.json'
with open(filename, 'r') as file:
  json_string = file.read()

print_json_attributes(json_string)
```
