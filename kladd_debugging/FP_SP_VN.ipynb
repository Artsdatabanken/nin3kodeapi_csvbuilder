```python
"""Fetching FP and SP from 'Oversettelse' column in variabel spreadsheet of exceldocument from teams"""
import pandas as pd
fpsp = pd.read_csv("fpsp.csv",usecols=['Oversettelse'], skip_blank_lines=False)
#display(fpsp.head(20))
fpsp['Oversettelse'] = fpsp['Oversettelse'].str.strip()#removing surrounding spaces
fpsp['FP'] = fpsp['Oversettelse'].str.split('').str[1]
fpsp.loc[fpsp['FP'].isin(['','n']), 'FP'] = 'ny'
fpsp['SP'] = ""
fpsp['SP'] = fpsp['Oversettelse'].str.split().str[-1]
fpsp.loc[fpsp['SP'].isin(['ny','n']), 'SP'] = ''
fpsp = fpsp.fillna('')
#creating a csv file with the new FP and SP columns
fpsp.to_csv("fp_sp_cols_for_master.csv",index=False, sep=";")
```
